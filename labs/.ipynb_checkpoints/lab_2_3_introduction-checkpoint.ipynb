# 2.3.1 Basic Commands

```python
# imports and setup
import numpy as np
from scipy.stats.stats import pearsonr
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
plt.style.use('seaborn')
```

```python
# array creation
x = np.array([1, 6, 2])
y = np.array([1, 4, 3])
len(x), len(y)
```

```python
# array operations
x + y
```

```python
# matrix creation
x = np.asmatrix(np.arange(1, 5).reshape(2, 2).transpose())
x
```

```python
#matrix operations
np.power(x, 2)
```

```python
# random normal distribution & correlation
x = np.random.normal(size=50)
y = x + np.random.normal(loc=50, scale=.1, size=50)
pearsonr(x, y)[0]
```

```python
# random seed and basic stat functions
np.random.seed(3)
y = np.random.normal(size=100)
y.mean(), y.var(), np.sqrt(y.var()), y.std()
```

# 2.3.2 Graphics

```python
x = np.random.normal(size=100)
y = np.random.normal(size=100)
p = sns.jointplot(x, y, kind='scatter')
p.set_axis_labels(xlabel='x axis', ylabel='y axis')
plt.show()
```

```python
# create a sequence of numbers
x = np.arange(1, 11)
x
```

```python
# linearly spaced numbers
x = np.linspace(-np.pi, np.pi, num=50)
x
```

```python
x = np.linspace(-np.pi, np.pi, num=50)
y = x

# simulating R outer function
def pf(a, b):
    return math.cos(b) / (1 + a**2)

f = np.empty((len(x), len(y)))
 
for i in range(len(x)):
    for j in range(len(y)):
        f[i,j] = pf(x[i], y[j])

        
# contour plot
cp = plt.contour(x, y, f, 45, cmap='viridis')
plt.clabel(cp, inline=1, fontsize=10)
plt.show()
```

```python
# contour 2
fa = (f - f.transpose())/2
cp = plt.contour(x, y, fa, 15, cmap='viridis')
plt.clabel(cp, inline=1, fontsize=10)
plt.show()
```

```python
# heatmap
cp = plt.contourf(x, y, fa, 15, cmap='viridis')
plt.clabel(cp, inline=1, fontsize=10)
plt.colorbar()
plt.show()
```

```python
# 3d perspective
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, fa, cmap='viridis')
ax.view_init(30, 100)
plt.show()
```
